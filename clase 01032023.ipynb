```python
def getDFDataTypes(dataset):

    continuas=[]
    discretas=[]
    categoricas=[]

    for columnName in dataset.columns.values:
        tipoDato=dataset[columnName].dtype
        if ((tipoDato=='int64') or (tipoDato=='float64')):
            if (len(dataset[columnName].unique())<=30):
                discretas.append(columnName)
            else:
                continuas.append(columnName)
        else:
            categoricas.append(columnName)
            
    return continuas,discretas,categoricas


```

```python
continuas,discretas,categoricas=getDFDataTypes(dataset)
```
